```python
import numpy as np
import matplotlib.pyplot as plt

# Define the MDOF system properties
m1 = 20      # mass of first degree of freedom
m2 = 20.0      # mass of second degree of freedom
m3 = 20      # mass of third degree of freedom
m4 = 20      # mass of fourth degree of freedom

k1 = 1.0     # stiffness of first degree of freedom
k2 = 1     # stiffness of second degree of freedom
k3 = 1     # stiffness of third degree of freedom
k4 = 1     # stiffness of fourth degree of freedom

c1 = 0.1      # damping of first degree of freedom
c2 = 0.1      # damping of second degree of freedom
c3 = 0.1     # damping of third degree of freedom
c4 = 0.1     # damping of fourth degree of freedom

# Define the earthquake wave
t_max = 20.0  # maximum time
dt = 0.01     # time step
t = np.arange(0, t_max, dt)    # time vector
f = 20   # frequency of earthquake wave
a_max = 1  # maximum acceleration
ai = a_max * np.sin(2 * np.pi * f * t)   # acceleration vector

# Define the Newmark method parameters
gamma = 0.5
beta = 0.25
a0 = np.zeros((4,))
v0 = np.zeros((4,))
d0 = np.array([0, 0, 0, 0])

# Calculate the response of the system
n = len(t)
d = np.zeros((n, 4))
v = np.zeros((n, 4))
a = np.zeros((n, 4))
d[0] = d0
v[0] = v0
# a[0] = 
for i in range(1, n):
    M = np.array([[m1, 0, 0, 0], [0, m2, 0, 0], [0, 0, m3, 0], [0, 0, 0, m4]])
    K = np.array([[k1+k2, -k2, 0, 0], [-k2, k2+k3, -k3, 0], [0, -k3, k3+k4, -k4], [0, 0, -k4, k4]])
    C = np.array([[c1+c2, -c2, 0, 0], [-c2, c2+c3, -c3, 0], [0, -c3, c3+c4, -c4], [0, 0, -c4, c4]])

    d_pred = d[i-1] + dt * v[i-1] + ((1 - 2 * beta) / 2) * dt ** 2 * a[i-1]
    v_pred = v[i-1] + (1 - gamma) * dt * a[i-1]
    a_pred = np.linalg.solve(M, (ai[i-1] - np.dot(C, v_pred) - np.dot(K, d_pred)) )
    
    d[i] = d_pred + beta * dt ** 2 * a_pred
    v[i] = v_pred + gamma * dt * a_pred
    a[i] = a_pred

    
# Plot the response of the system
plt.figure()
plt.plot(t, d[:, 0], label='d1')
plt.plot(t, d[:, 1], label='d2')
plt.plot(t, d[:, 2], label='d3')
plt.plot(t, d[:, 3], label='d4')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('MDOF System Response to Earthquake')
plt.legend()
plt.show()    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define system parameters
m = np.array([1, 1, 1, 1])*10  # masses of each degree of freedom
k = np.array([1, 1, 1, 1])*2467  # stiffness of each degree of freedom
c = np.array([1, 1, 1, 1])*10 # damping coefficient of each degree of freedom

# Define earthquake wave parameters
dt = 0.01  # time step
T = 10  # duration of earthquake wave
t = np.arange(0, T+dt, dt)  # time vector
a = 1  # acceleration amplitude
w = 2*np.pi*2  # frequency


df = pd.read_fwf('gorkha_thimi.dat',header = None,skiprows = 34,colspecs = 'infer',infer_nrows = 10000)
df = pd.read_fwf('gorkha_data/patan.dat',header = None,skiprows = 34,colspecs = 'infer',infer_nrows = 10000)
# df = pd.read_fwf('turkey.asc',header = None,skiprows = 64,colspecs = 'infer',infer_nrows = 10000)
columns = ['t','a_n','a_e','a_u']
# columns = ['a_n']
df.columns = columns
# df['t'] = np.arange(0,df.shape[0]*0.01,0.01)
df = df/100
df['t'] = df['t']*100
# t = df[df.a_n>1]

df = df[df.t>=15]
df['t'] = np.arange(0,df.shape[0]*0.01,0.01)

a = df['a_n'].to_numpy()
t = df['t'].to_numpy()
g_d = integrate(integrate(a))

time = np.arange(t[0],t[-1],1/60)
interpol = interpolate.interp1d(t,a)
a = interpol(time)
# print()
t = time

# baseObj=BaselineRemoval(g_d)
# g_d=baseObj.ModPoly(6)

# Define initial conditions
x0 = np.zeros(4)
v0 = np.zeros(4)
y0 = np.concatenate((x0, v0))

dt = 1/60
# Define function to calculate acceleration due to earthquake wave
def acceleration(t):
    
    return a[t]
#     return a*(np.sin(w*t)+np.sin(0.1*w*t))

# Define function to calculate the system's equations of motion
def equations_of_motion(y, t,i):
    x = y[:4]  # displacement vector
    v = y[4:]  # velocity vector
    a_e = acceleration(i)
    f = np.zeros(4)
    f[0] = -k[0]*x[0] - c[0]*v[0] + k[1]*(x[1] - x[0]) + c[1]*(v[1] - v[0]) + m[0]*a_e
    f[1] = -k[1]*(x[1] - x[0]) - c[1]*(v[1] - v[0]) + k[2]*(x[2] - x[1]) + c[2]*(v[2] - v[1]) + m[1]*a_e
    f[2] = -k[2]*(x[2] - x[1]) - c[2]*(v[2] - v[1]) + k[3]*(x[3] - x[2]) + c[3]*(v[3] - v[2]) + m[2]*a_e
    f[3] = -k[3]*(x[3] - x[2]) - c[3]*(v[3] - v[2]) + m[3]*a_e
    x_dot = v
    v_dot = f/m
    return np.concatenate((x_dot, v_dot))

# Solve equations of motion using RK4 method
def solve_motion():
    y = np.zeros((len(t), 8))
    y[0] = y0
    for i in range(len(t)-1):
        k1 = dt*equations_of_motion(y[i], t[i],i)
        k2 = dt*equations_of_motion(y[i] + 0.5*k1, t[i] + 0.5*dt,i)
        k3 = dt*equations_of_motion(y[i] + 0.5*k2, t[i] + 0.5*dt,i)
        k4 = dt*equations_of_motion(y[i] + k3, t[i] + dt,i)
        y[i+1] = y[i] + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
    return y

# Get displacement responses
y = solve_motion()
d = y
x1 = y[:, 0]
x2 = y[:, 1]
x3 = y[:, 2]
x4 = y[:, 3]

# Plot displacement responses
plt.figure(figsize=(8, 5))
plt.plot(t, x1, label='DOF 1')
plt.plot(t, x2, label='DOF 2')
plt.plot(t, x3, label='DOF 3')
plt.plot(t, x4, label='DOF 4')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('MDOF System Response to Earthquake Wave')
plt.legend()
plt.show()
```
